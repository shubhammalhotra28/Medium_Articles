```python
# 0-1 KNAPSACK:


'''
Given two integer arrays representing weights and profit of 'N' items, we need to find a subset of these items
which will give us maximum profit such that their cumulative weight is not more than a 'given number C'.
Each item can only be selected once, which means either we put an item in the knapsack or skip it
'''    


def knapsack(profits,weights,capacity):
    
    n = len(profits)
    
    def helper(currIndex,capacity):
        
        if currIndex >= n or currIndex < 0 or capacity <= 0:
            return 0
        
        
        ans1 = 0
        if weights[currIndex]<=capacity:
            # including the item
            ans1 = profits[currIndex] + helper(currIndex+1,capacity-weights[currIndex]) 
        
        # excuding the item
        ans2 = helper(currIndex+1,capacity)
    
        return max(ans1,ans2)

    return helper(0,capacity)



def knapsackMemo(profits,weights,capacity):
    
    n = len(profits)
    dp = [[-1 for j in range(capacity+1)]for i in range(n+1)]
    
    def helper(currIndex,capacity,dp):
        
        if currIndex >= n or currIndex < 0 or capacity <= 0:
            return 0
        
        
        if dp[currIndex][capacity] != -1:
            return dp[currIndex][capacity]
        
        ans1 = 0
        if weights[currIndex]<=capacity:
            # including the item
            ans1 = profits[currIndex] + helper(currIndex+1,capacity-weights[currIndex],dp) 
        
        # excuding the item
        ans2 = helper(currIndex+1,capacity,dp)
        
        dp[currIndex][capacity] = max(ans1,ans2)
    
        return dp[currIndex][capacity]

    return helper(0,capacity,dp)


# bottom up approach
def knapsackIterative(profits,weight,capacity):
    
    n = len(profits)
    # base case handled as well
    dp = [[0 for j in range(capacity+1)]for i in range(len(profits)+1)]
    
    for i in range(1,len(profits)+1):
        
        for j in range(1,capacity+1):
            
            ans1 = 0
            ans2 = 0
            
            if weights[i-1]<=j:
                # including 
                ans1 = profits[i-1]+dp[i-1][j-weights[i-1]]
            
            # excluding
            ans2 = dp[i-1][j]
            
            dp[i][j] = max(ans1,ans2)
            
    return dp[n][capacity]


profits = [1,6,10,16]
weights = [1,2,3,5]
capacity = 7

print(knapsackIterative(profits,weights,capacity))


print(knapsack(profits,weights,capacity))

print(knapsackMemo(profits,weights,capacity))
```

```python
# Unbounded knapsack


'''
Given two integer arrays represeting weights and rofits, we need to find a subset of these items which will 
give us maximum profit such that their cumulative weight is not more than a given number 'C'. We can 
assume an infinite supply of item quantities,each can be selected multiple times.


'''

def unboundedKnapsack(profits,weight,capacity):
    
    n = len(profits)
    
    def helper(currIndex,capacity):
        
        if capacity <= 0 or n==0 or currIndex >= n:
            return 0
        
        ans1 = 0
        
        if weight[currIndex] <= capacity:
                # including
                ans1 = profits[currIndex] + helper(currIndex,capacity-weight[currIndex])
        # excluding
        
        ans2 = helper(currIndex+1,capacity)
        
        return max(ans1,ans2)
    
    return helper(0,capacity)


def unboundedKnapsackMemo(profits,weight,capacity):
    
    n = len(profits)
    dp = [[-1 for j in range(capacity+1)]for i in range(n+1)]
    
    def helper(currIndex,capacity,dp):
        
        if capacity <= 0 or n==0 or currIndex >= n:
            return 0
        
        if dp[currIndex][capacity] != -1:
            return dp[currIndex][capacity]
        
        ans1 = 0
        
        if weight[currIndex] <= capacity:
                # including
                ans1 = profits[currIndex] + helper(currIndex,capacity-weight[currIndex],dp)
        # excluding
        
        ans2 = helper(currIndex+1,capacity,dp)
        
        dp[currIndex][capacity] = max(ans1,ans2)
        
        return dp[currIndex][capacity]
    
    return helper(0,capacity,dp)


def unboundedKnapsackI(profits,weight,capacity):
    
    n = len(profits)
    dp = [[0 for j in range(capacity+1)]for i in range(n+1)]
    
    for i in range(1,n+1):
        for j in range(1,capacity+1):
            ans1 = 0
            ans2 = 0
            if weight[i-1] <= j:
                
                ans1 = profits[i-1] + dp[i][j-weight[i-1]]
            
            ans2 = dp[i-1][j]
            
            dp[i][j] = max(ans1,ans2)
    
    return dp[n][capacity]
            








profits = [15,50,60,90]
weight = [1,3,4,5]
capacity = 8

print(unboundedKnapsack(profits,weight,capacity))

print(unboundedKnapsackMemo(profits,weight,capacity))
print(unboundedKnapsackI(profits,weight,capacity))
        
```
